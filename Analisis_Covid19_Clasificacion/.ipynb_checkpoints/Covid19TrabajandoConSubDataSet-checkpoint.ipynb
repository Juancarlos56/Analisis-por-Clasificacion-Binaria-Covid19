### Preprocesamiento

```python
#PREPROCESAMIENTO
# y va a contener la salida
salida = dataset['covid19_positive']
# X va a tener todos los datos, menos la etquet a de salida
df2 = dataset.drop('covid19_positive', 1)
```

```python
numbers = [0, 1, 6]
listCol = list(df2.columns)
val_numericas_cat_ord = listCol
aux = []
val_cat_nominales=[]
cont=0

for i in numbers:
    ten = val_numericas_cat_ord.pop(i-cont)
    cont= cont+1
    val_cat_nominales.insert(cont, ten)

print("Variables Categoricas Nominales \n\n",val_cat_nominales)
print("\nVariables numericas, y variables ordinales\n\n",val_numericas_cat_ord)
aux =val_numericas_cat_ord

#SEGUNDA FORMA DE PREPROCESAR

numeric_features = val_numericas_cat_ord
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_features = val_cat_nominales
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder())])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

clf = Pipeline(steps=[('preprocessor', preprocessor)])
#print(clf)

X = clf.fit_transform(df2)

#print(X)
#print(clf)

cnamesDataset1 = val_numericas_cat_ord
cnamesDataset2 = clf.named_steps['preprocessor'].transformers_[1][1]\
   .named_steps['onehot'].get_feature_names(categorical_features)


cnamesDataset1.extend(cnamesDataset2)
DatasetPreprocesado = pd.DataFrame(data=X,columns=cnamesDataset1)
DatasetPreprocesado = DatasetPreprocesado.assign(salida=salida.values)
DatasetPreprocesado.to_csv("DataSets/DatasetPreprocesadoNuevo.csv", sep=",",index = False) #sep es el separado, por defector es ","
DatasetPreprocesado.head()
```

### REDUCCION DE DIMENSIONALIDAD con 85% de la varianza 

```python
#Reducción de Dimensionalidad con PCA: un dataset con menores dimensiones necesita menos costo computacional (CPU) y memoria.
from sklearn.decomposition import PCA
X=DatasetPreprocesado# es la matriz de ratings.  
n_samples = X.shape[0]#cantidad de usuarios
n_features = X.shape[1]#cantidad de componentes. n_components must be between 0 and min(n_samples, n_features)
num_componentsSeleccionados=50#Para modificar el número de componentes

#Primera forma: para obtener el porcentaje de varianza obtenido con cada componente
print('********** Primera Forma')

num_components = min(n_samples, n_features)
num_components=num_componentsSeleccionados

#Se mantiene el 90% de la varianza de los datos
pca = PCA(0.85)
X_transformed = pca.fit_transform(X)#Esta es la matriz de componentes principales
eigenvalues = pca.explained_variance_
num_componentsSeleccionados = X_transformed.shape[1]

print('Varianzas:')
print(eigenvalues.round(3))

explained_variance_ratio_=pca.explained_variance_ratio_
print('Porcentaje de varianza de cada dimension con respecto a la varianza total:')
print(explained_variance_ratio_.round(3))

#Se obtiene el porcentaje acumulado de varianza
print('Porcentaje acumulado de varianza:')
explained_variance_ratio_cumsum=explained_variance_ratio_.cumsum()
print(explained_variance_ratio_cumsum.round(3))
```

```python
print("Numero de componentes seleccionados es: ", num_componentsSeleccionados)

num_pc= num_componentsSeleccionados

a = range(1,num_pc+1)
num_pc= a[::1]
num_pc


plt.plot(num_pc, explained_variance_ratio_, color='blue', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='orange', markersize=5, label = 'explained variance')
plt.plot(num_pc, explained_variance_ratio_cumsum, color='blue', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='orange', markersize=5, label = 'cumulative explained variance')
plt.ylim((0,1.2))
plt.title('Variance with '+str(num_componentsSeleccionados)+' principal components')
plt.xlabel('Principal Component ID')
plt.ylabel('Variance')
plt.legend(loc="upper right")
plt.show()

principalDf_grupo5 = pd.DataFrame(data = X_transformed
             , columns = num_pc)
principalDf_grupo5.describe()
```

### Clustering de subDataSet

```python
#Metodo Elbow
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn import metrics

X=principalDf_grupo5
K_range=range(1,20,1)
distortions=[]

for i in K_range:
    kmeanModel = KMeans(n_clusters=i,init='k-means++')
    kmeanModel.fit(principalDf_grupo5)
    distortions.append(sum(np.min(cdist(principalDf_grupo5, kmeanModel.cluster_centers_, 'euclidean'),axis=1)) / principalDf_grupo5.shape[0])

params = {'axes.labelsize': 'x-small'}
plt.rcParams.update(params)

fig1=plt.figure()
ex = fig1.add_subplot(111)
ex.plot(K_range, distortions, 'b*-')

plt.rc('font', size = 15)
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.savefig('Graficas/SubDataSet/GraficaDistortionsSubDataset.jpg', bbox_inches='tight', dpi=300)
```

### Tiempo de ejecuacion del cluster

```python
#Metodo Elbow
import time
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn import metrics

X=principalDf_grupo5
K_range=range(2,20,1)
distortions2=[]
tiempoTotal = []


def obtenerK2(distortions):
    total = distortions[0] - distortions[len(distortions)-1]
    p2p = []
    valoresCurvatura = []
    valoresCurvaturaX = []
    for i in range(0,len(distortions)):
        if (i<len(distortions)-1):
            dpp = distortions[i] - distortions[i+1] 
            p2p.append(dpp)
            porcentaje = (dpp*100)/total 
            #print(distortions[i], " | ", i+1," | ",dpp, " | ", porcentaje)
    
    #print("\n\nCambio de Curvatura:")
    for i in range(0,len(p2p)):
        if (i<len(distortions)-2):       
            if (p2p[i] < p2p[i+1] ):
                valoresCurvatura.append(p2p[i+1])
                valoresCurvaturaX.append((i+1)+1)
                #print((i+1)+1, " | ", p2p[i+1] )
                
    if(len(valoresCurvatura)>0): 
        #print("vc ",valoresCurvatura)
        valor = max(valoresCurvatura)
        numeroCluster = valoresCurvaturaX[valoresCurvatura.index(valor)]
        return (numeroCluster)
    


for i in K_range:
    start = time.time()
    kmeanModel = KMeans(n_clusters=i,init='k-means++')
    kmeanModel.fit(principalDf_grupo5)
    distortions2.append(sum(np.min(cdist(principalDf_grupo5, kmeanModel.cluster_centers_, 'euclidean'),axis=1)) / principalDf_grupo5.shape[0])  
    numeroCluster = obtenerK2(distortions2)
    print("Con ", i ," El Mejor K para este tipo de datos es: ", numeroCluster)
    end = time.time()    
    tiempo = end - start
    tiempoTotal.append(tiempo)

print("\nEl tiempo de Ejecusion en milisegundos")
print(tiempoTotal)
plt.figure()
plt.rcParams.update({'font.size': 10})
#plt.bar(range(len(grupo)), valores, color='red')
plt.plot(K_range, tiempoTotal, color='orange', marker='o')
##Pone el valor de los indices en el eje x
#plt.xticks(range(len(grupo)), grupo, rotation='vertical')
plt.title("Tiempo de Ejecucion segun el numero de Clusters")
plt.xlabel("Cluster")
plt.ylabel("Milesegundos")
plt.savefig('Graficas/SubDataSet/GraficaTiempoEjecucionClusterSubDataset2.jpg', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import axes3d
from sklearn import datasets
from sklearn.cluster import KMeans
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

num_components=3
pca = PCA(num_components)
principalComponents = pca.fit_transform(principalDf_grupo5)

#The explained variance tells you how much information (variance) can be attributed to each of the principal components. This is important as while you can convert 4 dimensional space to 2 dimensional space, you lose some of the variance (information) when you do this. By using the attribute explained_variance_ratio_, you can see that the first principal component contains 72.77% of the variance and the second principal component contains 23.03% of the variance. Together, the two components contain 95.80% of the information.
explained_variance=pca.explained_variance_ratio_
explained_variance

a = range(num_components)
num_pc= a[::1]

principalDf = pd.DataFrame(data = principalComponents
             , columns = num_pc)

principalDf=round(principalDf, 2)
explained_variance
a = range(num_components)
num_pc= np.linspace(1,num_components,num_components)

#K means Clustering 
def doKmeans(X, nclust=2):
    model = KMeans(nclust)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = doKmeans(principalDf, 8)
kmeans = pd.DataFrame(clust_labels)
finalDf = pd.concat([principalDf, kmeans], axis = 1)
finalDf.columns = ['ComponentePrincipal1', 'ComponentePrincipal2', 'ComponentePrincipal3', 'target']
```

```python
y = DatasetPreprocesado2['salida']
# X va a tener todos los datos, menos la etquet a de salida
X = DatasetPreprocesado2.drop('salida', 1)
```

```python
#PROCESO PCA CON N_COMPONENTS. LA MAYOR CANTIDAD DE COMPONENTES NO GARANTIZA UN MEJOR RESULTADO
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
print("Tamaño de X train: ", X_train.shape, "Tamaño de Y train: ", y_train.shape)
print("Tamaño de X test: ", X_test.shape, "Tamaño de Y test: ", y_test.shape)

#pca = PCA()  
accuracy = []
f1 = []
recall = []
precision = []

#num_comp = len(dataset)
num_comp = 40
for i in range(1,num_comp):
    
    pca = PCA(i)
    X_train_comp = pca.fit_transform(X_train)  
    X_test_comp = pca.transform(X_test) 
    explained_variance = pca.explained_variance_ratio_  
    classifier = RandomForestClassifier(max_depth=2, random_state=0)  
    classifier.fit(X_train_comp, y_train)
    # Predicting the Test set results
    y_pred_comp = classifier.predict(X_test_comp) 
    cm = confusion_matrix(y_test, y_pred_comp)  
    accuracy.append(accuracy_score(y_test, y_pred_comp))
    f1.append(f1_score(y_test, y_pred_comp, average='weighted'))
    recall.append(recall_score(y_test, y_pred_comp, average='weighted'))
    precision.append(precision_score(y_test, y_pred_comp, average='weighted'))
    #print ('Accuracy:', accuracy_score(y_test, y_pred_comp))
```

```python
ejeX = np.linspace(1, num_comp, num_comp-1, endpoint=False) 

params = {'legend.fontsize': 'x-small',
            'figure.figsize': (9,6),
            'axes.labelsize': 'xx-small',
            'axes.titlesize':'xx-small',
            'xtick.labelsize':'xx-small',
            'ytick.labelsize':'xx-small'}
plt.rcParams.update(params)
plt.figure()
# Dibuja el gráfico
plt.plot(ejeX, accuracy, color='lightblue', linestyle='dashed',linewidth=2,marker='o',markerfacecolor='green',markersize=5)   
plt.title("Accuracy")   # Establece el título del gráfico
plt.xlabel("Numero de Componentes")   # Establece el título del eje x
plt.ylabel("Porcentaje de Presión")   # Establece el título del eje y
plt.grid(True)
plt.savefig('Graficas/SubDataSet/Grafica_Accuracy.jpg', bbox_inches='tight', dpi=300)
```

```python
print('Mejor Accuracy:',max(accuracy),' con un numero de componente de: ',accuracy.index(max(accuracy))+1)
print('Mejor F1:',max(f1),' con un numero de componente de: ',f1.index(max(f1))+1)
print('Mejor Recall:',max(recall),' con un numero de componente de: ',recall.index(max(recall))+1)
print('Mejor Precision:',max(precision),' con un numero de componente de: ',precision.index(max(precision))+1)
```
