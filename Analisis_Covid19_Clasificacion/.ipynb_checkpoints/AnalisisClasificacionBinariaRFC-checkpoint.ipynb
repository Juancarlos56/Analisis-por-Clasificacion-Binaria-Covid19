#### CAMBIO DE VALORES A FORMA NUMERICA

```python
df['region'] = df['region'].fillna("NR")

df['age'] = df['age'].fillna(0)
df['age'] = df['age'].replace({"30_40":0,"0_10":1,"10_20":2,"20_30":3,"30_40":4,"40_50":5,"50_60":6,"60_70":7,"70_80":8,"80_90":9,"90_100":10,"100_110":11})

df['height'] = df['height'].fillna(round(np.mean(df['height']),0))

df['weight'] = df['weight'].fillna(round(np.mean(df['weight']),0))

df['bmi'] = df['bmi'].fillna(round(df["bmi"].mean(),0))

df['smoking'] = df['smoking'].fillna(-1)
df['smoking'] = df['smoking'].replace({"never":0,"quit10":1,"quit5":2,"quit0":3,"vape":4,"yeslight":5,"yesmedium":6,"yesheavy":7})

df['alcohol'] = df['alcohol'].fillna(0)

df['cannabis'] = df['cannabis'].fillna(-1)

df['amphetamines'] = df['amphetamines'].fillna(-1)

df['cocaine'] = df['cocaine'].fillna(-1)

df['lsd'] = df['lsd'].fillna(-1)

df['mdma'] = df['mdma'].fillna(-1)

df['contacts_count'] = df['contacts_count'].fillna(7)

df['house_count'] = df['house_count'].fillna(3)

df['working'] = df['working'].fillna('vacio')
df['working'] = df['working'].replace({"vacio":0,"home":1,"never":2,"stopped":3,"travel critical":4,"travel non critical":5})

df['rate_reducing_mask'] = df['rate_reducing_mask'].fillna(0)

df['opinion_infection'] = df['opinion_infection'].fillna(36)

df['opinion_mortality'] = df['opinion_mortality'].fillna(21)

df['risk_infection'] = df['risk_infection'].fillna(11.4)

df['risk_mortality'] = df['risk_mortality'].fillna(0.34)
```

#### PREPROCESAMIENTO 

```python
numbers = [0, 1, 6]
listCol = list(df2.columns)
val_numericas_cat_ord = listCol
aux = []
val_cat_nominales=[]
cont=0

for i in numbers:
    ten = val_numericas_cat_ord.pop(i-cont)
    cont= cont+1
    val_cat_nominales.insert(cont, ten)

print("Variables Categoricas Nominales \n\n",val_cat_nominales)
print("\nVariables numericas, y variables ordinales\n\n",val_numericas_cat_ord)
aux =val_numericas_cat_ord

#SEGUNDA FORMA DE PREPROCESAR

numeric_features = val_numericas_cat_ord
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_features = val_cat_nominales
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder())])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

clf = Pipeline(steps=[('preprocessor', preprocessor)])
#print(clf)

X = clf.fit_transform(df2)

#print(X)
#print(clf)

cnamesDataset1 = val_numericas_cat_ord
cnamesDataset2 = clf.named_steps['preprocessor'].transformers_[1][1]\
   .named_steps['onehot'].get_feature_names(categorical_features)


cnamesDataset1.extend(cnamesDataset2)
DatasetPreprocesado2 = pd.DataFrame(data=X,columns=cnamesDataset1)
DatasetPreprocesado2 = DatasetPreprocesado2.assign(salida=salida.values)
DatasetPreprocesado2.to_csv("DataSets/DatasetPreprocesado.csv", sep=",",index = False) #sep es el separado, por defector es ","
DatasetPreprocesado2.head()
```

### REDUCCION DE DIMENSIONALIDAD sin especificar el numero de componentes

```python
#Reducción de Dimensionalidad con PCA: un dataset con menores dimensiones necesita menos costo computacional (CPU) y memoria.
from sklearn.decomposition import PCA
X=df3# es la matriz de ratings. 
n_samples = X.shape[0]#cantidad de usuarios
n_features = X.shape[1]#cantidad de componentes. n_components must be between 0 and min(n_samples, n_features)
num_componentsSeleccionados=50#Para modificar el número de componentes

#Primera forma: para obtener el porcentaje de varianza obtenido con cada componente
print('********** Primera Forma')

num_components = min(n_samples, n_features)
num_components=num_componentsSeleccionados

#Se mantiene el 90% de la varianza de los datos
pca = PCA(0.85)
X_transformed = pca.fit_transform(X)#Esta es la matriz de componentes principales
eigenvalues = pca.explained_variance_
num_componentsSeleccionados = X_transformed.shape[1]

print('Varianzas:')
print(eigenvalues.round(3))

explained_variance_ratio_=pca.explained_variance_ratio_
print('Porcentaje de varianza de cada dimension con respecto a la varianza total:')
print(explained_variance_ratio_.round(3))

#Se obtiene el porcentaje acumulado de varianza
print('Porcentaje acumulado de varianza:')
explained_variance_ratio_cumsum=explained_variance_ratio_.cumsum()
print(explained_variance_ratio_cumsum.round(3))
```

```python
print("Numero de componentes seleccionados es: ", num_componentsSeleccionados)
num_pc= num_componentsSeleccionados

a = range(1,num_pc+1)
num_pc= a[::1]


plt.figure(figsize=(14,10))
plt.subplot(2, 2, 1)

plt.plot(num_pc, explained_variance_ratio_, color='blue', linestyle='dashed', linewidth = 1, 
         marker='o', markerfacecolor='orange', markersize=5, label = 'varianza explicada')
#plt.title('Variance with '+str(num_componentsSeleccionados)+' principal components')
plt.xlabel('Componentes Principales ID')
plt.ylabel('Varianza')
plt.legend(loc="upper right")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(num_pc, explained_variance_ratio_cumsum, color='blue', linestyle='dashed', linewidth = 1, 
         marker='o', markerfacecolor='orange', markersize=5, label = 'varianza explicativa acumulativa')
#plt.title('Variance with '+str(num_componentsSeleccionados)+' principal components')
plt.xlabel('Componentes Principales ID')
plt.ylabel('Varianza')
plt.legend(loc="upper right")
plt.grid(True)
plt.savefig('Graficas/DataSetCompleto/DataSetCompleto_Varianza_VarianzaAcumulada.jpg', bbox_inches='tight', dpi=300)
plt.show()

principalDfR = pd.DataFrame(data = X_transformed
             , columns = num_pc)
principalDfR.describe()
```

```python
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import axes3d
from sklearn import datasets
from sklearn.cluster import KMeans
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

num_components=3
pca = PCA(num_components)
principalComponents = pca.fit_transform(principalDfR)

#The explained variance tells you how much information (variance) can be attributed to each of the principal components. This is important as while you can convert 4 dimensional space to 2 dimensional space, you lose some of the variance (information) when you do this. By using the attribute explained_variance_ratio_, you can see that the first principal component contains 72.77% of the variance and the second principal component contains 23.03% of the variance. Together, the two components contain 95.80% of the information.
explained_variance=pca.explained_variance_ratio_
explained_variance

a = range(num_components)
num_pc= a[::1]

principalDf = pd.DataFrame(data = principalComponents
             , columns = num_pc)

principalDf=round(principalDf, 2)
explained_variance
a = range(num_components)
num_pc= np.linspace(1,num_components,num_components)

#K means Clustering 
def doKmeans(X, nclust=2):
    model = KMeans(nclust)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = doKmeans(principalDf, 8)

kmeans = pd.DataFrame(clust_labels)
finalDf = pd.concat([principalDf, kmeans], axis = 1)
finalDf.columns = ['ComponentePrincipal1', 'ComponentePrincipal2', 'ComponentePrincipal3', 'target']


from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from IPython.display import HTML
import matplotlib.animation as animation

fig = plt.figure(figsize = (10,12))
ax1 = fig.add_subplot(111, projection='3d')
ax1.set_xlabel('Componente Principal 1', fontsize=10)
ax1.set_ylabel('Componente Principal 2', fontsize=10)
ax1.set_zlabel('Componente Principal 3', fontsize=10)
ax1.set_title('PCA 3D',fontsize=15)
targets=[0,1,2,3,4,5,6,7]
colors=['blue','green','red', 'y','purple','cyan','pink','orange']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax1.scatter(finalDf.loc[indicesToKeep, 'ComponentePrincipal1']
               , finalDf.loc[indicesToKeep, 'ComponentePrincipal2']
               , finalDf.loc[indicesToKeep, 'ComponentePrincipal3']
               , c=color
               )

ax1.legend(["Grupo 0","Grupo 1","Grupo 2","Grupo 3","Grupo 4","Grupo 5","Grupo 6","Grupo 7"])
ax1.grid()
plt.savefig('Graficas/DataSetCompleto/Grafica3D_CLUSTER.jpg', bbox_inches='tight', dpi=300)
plt.show()
```

```python
#PROCESO PCA CON N_COMPONENTS. LA MAYOR CANTIDAD DE COMPONENTES NO GARANTIZA UN MEJOR RESULTADO
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10) 
print("Tamaño de X train: ", X_train.shape, "Tamaño de Y train: ", y_train.shape)
print("Tamaño de X test: ", X_test.shape, "Tamaño de Y test: ", y_test.shape)

#pca = PCA()  
accuracy = []
f1 = []
recall = []
precision = []

#num_comp = len(dataset)
num_comp = 50
for i in range(1,num_comp):
    
    pca = PCA(i)
    X_train_comp = pca.fit_transform(X_train)  
    X_test_comp = pca.transform(X_test) 
    explained_variance = pca.explained_variance_ratio_  
    classifier = RandomForestClassifier(max_depth=2, random_state=0)  
    classifier.fit(X_train_comp, y_train)
    # Predicting the Test set results
    y_pred_comp = classifier.predict(X_test_comp) 
    cm = confusion_matrix(y_test, y_pred_comp)  
    accuracy.append(accuracy_score(y_test, y_pred_comp))
    f1.append(f1_score(y_test, y_pred_comp, average='weighted'))
    recall.append(recall_score(y_test, y_pred_comp, average='weighted'))
    precision.append(precision_score(y_test, y_pred_comp, average='weighted'))
    #print ('Accuracy:', accuracy_score(y_test, y_pred_comp))
```

```python
ejeX = np.linspace(1, num_comp, num_comp-1, endpoint=False) 
plt.figure()
# Dibuja el gráfico
plt.plot(ejeX, accuracy, color='lightblue', linestyle='dashed',linewidth=2,marker='o',markerfacecolor='green',markersize=5)   
plt.title("Accuracy")   # Establece el título del gráfico
plt.xlabel("Numero de Componentes")   # Establece el título del eje x
plt.ylabel("Porcentaje de Presión")   # Establece el título del eje y
plt.grid(True)
plt.savefig('Graficas/SubDataSet/Grafica_Accuracy.jpg', bbox_inches='tight', dpi=300)
```

```python
def get_redundant_pairs(dfPre):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = dfPre.columns
    for i in range(0, dfPre.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(dfPre, n=5):
    au_corr = dfPre.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(dfPre)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```
